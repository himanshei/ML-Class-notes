```python
import numpy as np
```

```python
#plot()

x=np.linspace(0,10,100)
y=np.sin(x)
fig=plt.figure()
ax=plt.axes()
ax.plot(x,y)
```

```python
x=np.linspace(0,10,100)
plt.plot(x,x+1,linestyle='solid')
plt.plot(x,x+2,linestyle='dotted')
plt.plot(x,x+3,linestyle='dashdot')
plt.plot(x,x+4,linestyle='dashed')
```

```python
x=np.linspace(0,10,100)
plt.plot(x,x+1,linestyle='-')
plt.plot(x,x+2,linestyle='--')
plt.plot(x,x+3,linestyle='-.')
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x))
plt.xlabel('X Values')
plt.ylabel('Sine of x')
plt.title('x vs sin(x)')
```

```python
#Legend

x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),label='Sinx')
plt.plot(x,np.cos(x),label='Cosx')
plt.legend(loc='center')

#best,upperright,upperleft,lowerleft,lowerright,right, center left,center right, upper center,center
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),label='Sinx')
plt.plot(x,np.cos(x),label='Cosx')
plt.legend(loc='upper left')
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),label='Sinx')
plt.plot(x,np.cos(x),label='Cosx')
plt.plot(x,np.tan(x),label='Tanx')
plt.legend(loc='best')
```
